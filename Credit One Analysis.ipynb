```python
# Distribution of age, with an overlay of a density plot
age = credit['AGE']
age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Customers' Ages")
```

```python
age = credit['LIMIT_BAL']
age_dist = sns.distplot(age)
age_dist.set_title("Distribution of Customers' Limit")
```

```python
fig = sns.FacetGrid(credit, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by Sex')
fig.add_legend()

#more younger females, more older males
```

```python
fig = sns.FacetGrid(credit, hue='SEX', aspect=4)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit['LIMIT_BAL'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Limit Bal Grouped by Sex')
fig.add_legend()

#more males with a smaller limit, more females with a slightly larger limit
```

```python
fig = sns.FacetGrid(credit, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'AGE', shade=True)
oldest = credit['AGE'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of Age Grouped by defaulting')
fig.add_legend()
```

```python
fig = sns.FacetGrid(credit, hue='default payment next month', aspect=4)
fig.map(sns.kdeplot, 'LIMIT_BAL', shade=True)
oldest = credit['LIMIT_BAL'].max()
fig.set(xlim=(0,oldest))
fig.set(title='Distribution of limit bal Grouped by defaulting')
fig.add_legend()

#greater number of defaults with a low limit
```

```python
#LINEAR PLOT age with defaulting
sns.lmplot('AGE', 'default payment next month', data=credit)
#there does not appear to be much here, chance of deafulting based on just age seems to slightly increase as age increases.
```

```python
#LINEAR PLOT age with defaulting & SEX
sns.lmplot('AGE', 'default payment next month', data=credit, hue='SEX')
#This shows that the chance of defaulting increases slighlty as men age, whereas the chance of women defaulting stay the same.
```

```python
#LINEAR PLOT limit bal with defaulting
sns.lmplot('LIMIT_BAL', 'default payment next month', data=credit)

#looks like this shows the higher the limit balance, the less likely to default. 
#Which could mean to customers that have high limit balances, tend to have better credit scores and are more fianancially responsible
```
